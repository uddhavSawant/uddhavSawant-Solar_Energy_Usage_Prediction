```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['global-data-on-sustainable-energy (1).csv']))
print(df)
```

```python
df.describe()
```

```python
df.head()
```

```python
target = "Primary energy consumption per capita (kWh/person)"
```

```python
features = df.select_dtypes(include=[np.number]).drop(columns=["Year", target], errors='ignore')
```

```python
imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features_scaled, df[target].dropna(), test_size=0.2, random_state=42
)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R Score: {r2}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.axline([0, 0], [1, 1], color="red", linestyle="--")
plt.show()


residuals = y_test - y_pred
plt.figure(figsize=(8, 6))
sns.histplot(residuals, bins=30, kde=True)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Residual Distribution")
plt.show()
```
